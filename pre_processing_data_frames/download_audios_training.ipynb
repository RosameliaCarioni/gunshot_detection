## Reading from "other" file, which contains sounds no generated by gunshots

```python
other = pd.read_csv('data/other/ecoGuns_SST_Guns_Training-OtherSounds.txt', delimiter = "\t")
other = other.drop (columns=['View', 'Channel', 'Begin Path', 'site', 'uniqueID', 'Notes', 'Score final', 'ID even', 'type shot', 'total shots', 'Tags', 'Final'])
other['delta time (s)'] =  other['End Time (s)'] - other['Begin Time (s)']
```

```python
other_distinct = pd.unique(other['Begin File'])
```

```python
def get_amazon_sound_file_other (file_name):
    import requests
    site = file_name.split('_')[0]
    doc = requests.get(f'https://congo8khz-pnnn.s3.us-west-2.amazonaws.com/recordings/wav/{site}/{file_name}')
    with open(f"data/aws_other/{file_name}", 'wb') as f:
        f.write(doc.content)

files_not_aws = set()

for file in other_distinct:
    try:
        get_amazon_sound_file_other(file)
    except Exception:
        files_not_aws.add(file)
```
