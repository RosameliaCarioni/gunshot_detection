```python
import pandas as pd
from pydub import AudioSegment
```

# 1. Re naming of columns in df and cliping of audio files

```python
# Pass time in seconds
def clip_audio_and_save (instance_of_df, extra_name):
    time_start = instance_of_df['File Offset (s)']
    time_duration = instance_of_df['Delta Time (s)']
    file_location = 'data/Sounds/' + instance_of_df['File Name']
    file_name = 'data/Clips/' + extra_name + str(instance_of_df.name) + '.wav'
    second_to_milliseconds = 1000
    audio = AudioSegment.from_wav(file_location)
    clip = audio[time_start*second_to_milliseconds: time_start*second_to_milliseconds + time_duration*second_to_milliseconds]
    clip.export(file_name, format='wav')
    return file_name

def clip_all_df (df, extra_name):
    df['Clip Location'] = df.apply(clip_audio_and_save, axis = 1, extra_name = extra_name)
```


## Other: non gunshots

```python
other = pd.read_csv('data/other/ecoGuns_SST_Guns_Training-OtherSounds.txt', delimiter="\t")
other = other.drop(
    columns=['View', 'Channel', 'Begin Path', 'site', 'uniqueID', 'Notes', 'Score final', 'ID even', 'type shot',
             'total shots', 'Tags', 'Final', 'Selection'])
other['Delta Time (s)'] = other['End Time (s)'] - other['Begin Time (s)']
other = other.rename(columns= {'Relative Path': 'site', 'Begin File': 'File Name'})
other
```

```python
clip_all_df(other, 'other')
other.to_csv('data_frames/ecoGuns_SST_Guns_Training-OtherSounds.csv')
```

From this, I have in total 94  clips,
with a total of 616.36 seconds,
approximately 10 minutes

## Ecoguns: gunshots




```python
ecoguns = pd.read_csv('data/ecoguns/Guns_Training_ecoGuns_SST.txt', delimiter = "\t")
ecoguns = ecoguns.drop(columns= ['view', 'channel', 'begin path', 'duration (s)', 'time', 'distance(m)', 'calibre', 'angle', 'elevation', 'uniqueID', 'date', 'Selection', 'site.1'])
ecoguns['Delta Time (s)'] =  ecoguns['end time'] - ecoguns['begin time']
ecoguns = ecoguns.rename(columns={'begin time': 'Begin Time (s)', 'end time': 'End Time (s)', 'low freq': 'Low Freq (Hz)', 'high freq': 'High Freq (Hz)', 'Begin File': 'File Name'})
ecoguns
```
