# This notebook reads in the originally given .txt files that contain information about the test data, and deletes unnecesary columns from the dataframes. It also calculates the delta time which is used in the extraction of the clips from the long audio files. 

```python
pnnn_1 = pd.read_csv('data/data_frame/test/pnnn/nn_201804_apr_selected days-HB-guns.txt', delimiter = "\t")
pnnn_2 = pd.read_csv('data/data_frame/test/pnnn/nn01g_20180702_000000_00-24hr_hb-guns.txt', delimiter = "\t")
pnnn_3 = pd.read_csv('data/data_frame/test/pnnn/nn06d_20180311_000000.guns-hb-possible.txt', delimiter = "\t")
pnnn_4 = pd.read_csv('data/data_frame/test/pnnn/nn201710_hbguns_10days_gunsOnly.txt', delimiter = "\t")

korup = pd.read_csv('data/data_frame/test/korup/Korup_4kHz_Gunshots_Merged.txt', delimiter= "\t")
```

## Pnnn_1 

```python
pnnn_1 = pnnn_1.drop(columns= ['View', 'Channel', 'Begin Path', 'Selection', 'File Offset (s)', 'truepos', 'score-guns', 'analyst' ])
pnnn_1['delta time (s)'] =  pnnn_1['End Time (s)'] - pnnn_1['Begin Time (s)']
```

```python
pnnn_1.to_csv('data_frames/test/pnnn_1.csv')
```
